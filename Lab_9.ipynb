# Shortest Path Using Dijkstra Algorithm in Python
Problem Description:
Given a weighted graph with non negative edge weights. Find the shortest distance from a source vertex to all other vertices using Dijkstra algorithm

```python
import heapq

def dijkstra(graph, start):
    distances = {node: float('inf') for node in graph}
    previous = {node: None for node in graph}
    distances[start] = 0
    priority_queue = [(0, start)]

    while priority_queue:
        current_distance, current_node = heapq.heappop(priority_queue)

        if current_distance > distances[current_node]:
            continue

        for neighbor, weight in graph[current_node].items():
            new_distance = current_distance + weight
            if new_distance < distances[neighbor]:
                distances[neighbor] = new_distance
                previous[neighbor] = current_node
                heapq.heappush(priority_queue, (new_distance, neighbor))
    return distances, previous

def get_path(previous, start, end):
    path = []
    current = end
    while current is not None:
        path.append(current)
        current = previous[current]
    path.reverse()

    if path and path[0] == start:
        return path
    else:
        return None

graph = {}
try:
    while True:
        n_input = input("Enter number of nodes: ").strip()
        try:
            n = int(n_input)
            break
        except ValueError:
            print("Please enter a valid integer for the number of nodes.")

    print("Enter nodes:")
    for _ in range(n):
        node = input().strip()
        if node:
            graph[node] = {}

    print("\nEnter edges (format: node1 node2 weight). Type 'done' to stop:")
    while True:
        entry = input().strip()
        if entry.lower() == 'done' or not entry:
            break
        try:
            parts = entry.split()
            if len(parts) != 3:
                raise ValueError("Incorrect number of values")
            u, v, w = parts[0], parts[1], int(parts[2])
            if u not in graph: graph[u] = {}
            if v not in graph: graph[v] = {}
            graph[u][v] = w
            graph[v][u] = w
        except ValueError:
            print("Invalid format. Use: node1 node2 weight (e.g., A B 5)")

    while True:
        start_node = input("\nEnter source node (or 'exit' to quit): ").strip()
        if start_node.lower() == 'exit':
            break
        if start_node not in graph:
            print(f"Node '{start_node}' not found.")
            continue

        distances, previous = dijkstra(graph, start_node)
        print(f"\nShortest paths from {start_node}:")
        for node in graph:
            path = get_path(previous, start_node, node)
            if path:
                print(f"{start_node} -> {node} = {distances[node]}, Path: {' -> '.join(path)}")
            else:
                print(f"{start_node} -> {node} = Unreachable")
except EOFError:
    print("Input stream closed.")
```

#Theory
Dijkstra algorithm finds shortest distance from one source node to all other nodes in a weighted graph with non negative weights. Dictionary structure stores nodes as keys and neighbor nodes with weights as values.

##Algorithm Steps

1. Represent graph using dictionary.
2. Set distance of every node as infinity.
3. Set distance of source node as 0.
4. Select node with smallest distance among unvisited nodes.
5. Update distance of neighbor nodes.
6. Repeat until all nodes get visited.

```python
def dijkstra(graph, start):

    distances = {}
    visited = []

    for node in graph:
        distances[node] = float('inf')

    distances[start] = 0

    while len(visited) < len(graph):

        min_node = None
        min_distance = float('inf')

        for node in graph:
            if node not in visited and distances[node] < min_distance:
                min_distance = distances[node]
                min_node = node

        visited.append(min_node)

        for neighbor, weight in graph[min_node].items():
            new_distance = distances[min_node] + weight

            if new_distance < distances[neighbor]:
                distances[neighbor] = new_distance

    return distances
```

```python
graph = {
'A': {'B': 4, 'C': 2},
'B': {'A': 4, 'C': 5, 'D': 10},
'C': {'A': 2, 'B': 5, 'D': 3},
'D': {'B': 10, 'C': 3}
}

result = dijkstra(graph, 'A')

for node in result:
    print("Distance from A to", node, "=", result[node])
```
