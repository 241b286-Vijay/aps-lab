```python
# Exercise 4
def func4(a, b):
    b.sort() # Time complexity for this function is O(Nlog N) where N = len(b)
    intersect = 0
    for x in a:
        if binary_search(b, x) >= 0:
            intersect = intersect + 1
    return intersect

def binary_search(arr, target):
    low, high = 0, len(arr) - 1
    while low <= high:
        mid = (low + high) // 2
        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            low = mid + 1
        else:
            high = mid - 1
    return -1
```

```python
# Exercise 6
def func8(A):

    func8_helper(A, 0, len(A) -1)
    return A

def func8_helper(A, i, j):

    n = j - i + 1

    if n <= 1:
        return


    if n == 2:
        if A[i] > A[j]:
            A[i], A[j] = A[j], A[i]
        return

    m = n // 3


    func8_helper(A, i, j - m)

    func8_helper(A, i + m, j)

    func8_helper(A, i, j - m)

```

```python
# write a function named as sum_arr to find sum of elements in given array
def sum(arr):
  sum=0
  for i in range(len(arr)):
    sum=sum+arr[i]
  return sum
```

```python
#call the function
import time
start=time.time()
print(sum(arr=[12,1234,45,67,1]))
end=time.time()
print(end-start)
#print(sum_arr)
```

```python
#write a function named as max_min to find minimum (or maximum) element in an array
def max_min(arr):
  max=arr[0]
  min=arr[0]
  for i in range(len(arr)):
    if arr[i]>max:
     max=arr[i]
    elif arr[i]<min:
      min=arr[i]
```

```python
# write a function named as Counting_EvenOdd to count number of even and odd elements in an array
def Counting_EvenOdd(num):
  even=0
  odd=0
  for i in range(len(num)):
    if num[i]%2==0:
      even+=1
    elif num[i]%2!=0:
      odd+=1
      return even,odd
```

```python
import time
start=time.time()
print(Counting_EvenOdd(num=[12,1234,45,67,1]))
end=time.time()
print(end-start)
```

```python
# write a function named as linear_Search to perform a linear search in 1D-array
def linear_Search(arr,x):
  for i in range(len(arr)):
    if arr[i]==x:
      return i
  return -1
```

```python
# call the function liner_Search for array = [1, 3, 5, 7, 9]
import time
start=time.time()
print(linear_Search(arr=[1, 3, 5, 7, 9],x=9))
end=time.time()
print(end-start)
```

```python
# write a function named as Binary_Search to perform binary search in 1D-array
def Binary_Search(arr,x):
  low=0
  high=len(arr)-1
  while low <= high:
    mid = (low + high) // 2
    if arr[mid] == x:
      return mid
    elif arr[mid] < x:
      low = mid + 1
    else:
      high = mid - 1
  return -1
```

```python
# call the function Binary_Search for array arr = [1, 3, 5, 7, 9,15,16,14,45]
import time
start=time.time()
print(Binary_Search(arr=[1, 3, 5, 7, 9, 14, 15, 16, 45],x=9)) # Changed array to sorted for Binary Search
end=time.time()
print(end-start)
```
